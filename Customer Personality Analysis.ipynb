```python
#detect outlier
features = ['Year_Birth','Income','Kidhome','Teenhome','Recency','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds','NumDealsPurchases','NumWebPurchases','NumCatalogPurchases','NumStorePurchases','NumWebVisitsMonth']
for i in features:
    plt.boxplot(df[i])
    plt.title(f"{df[i]}")
    plt.show()
```

the features that shows the amount has more number of outliers

```python
import datetime
date = []
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
print(f"max in dates = {max(df['Dt_Customer'])}")
print(f"min in dates = {min(df['Dt_Customer'])}")
# how many days the customer has been purchasing till now
for i in df['Dt_Customer']:
    i = i.date()
    date.append(i)

tdays = []
maxdate = max(date)
for i in date:
    j = maxdate - i
    tdays.append(j)
df['total_days'] = tdays
df
print(f"max no. of days: {max(df['total_days'])}")
print(f"min no. of days: {min(df['total_days'])}")
```

```python
print(df['Marital_Status'].value_counts())
print(df['Education'].value_counts())

df['lives_with'] = df['Marital_Status'].replace({
    "Married":"WithPartner",
    "Together":"WithPartner",
    "Single":"withoutPartner",
    "Divorced":"withoutPartner",
    "Widow":"withoutPartner",
    "Alone":"withoutPartner",
    "Absurd":"withoutPartner",
    "YOLO":"withoutPartner"
})
df['Education_s'] = df['Education'].replace({
    "Basic":"UG",
    "2n Cycle":"UG",
    "Graduation":"G",
    "Master":"PG",
    "PhD":"PG"
})
```

```python
#customers age

df['Age'] = 2023 - df['Year_Birth']
df["Spent"] = df["MntWines"] + df["MntFruits"] + df["MntMeatProducts"] + df["MntFishProducts"] + df["MntSweetProducts"] + df["MntGoldProds"]
df['Family_size'] = df['Kidhome'] + df['Teenhome'] + df['lives_with'].replace({"withoutPartner":1,"WithPartner":2})
df["Children"] = df["Kidhome"] + df["Teenhome"]
df["Parent"] = np.where(df.Children > 0, 1, 0)
#now drop all redundant features
df = df.drop(['ID','Year_Birth','Education','Marital_Status','Dt_Customer', 'Z_CostContact','Z_Revenue'],axis = 1)
df
```

# Data Preprocessing

```python
df['total_days'] = df['total_days'].astype(str).str.replace("days", "").str.strip()
df['total_days'] = df['total_days'].astype(int)
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df['lives_with'] = lb.fit_transform(df['lives_with'])
df['Education_s'] = lb.fit_transform(df['Education_s'])
df.info()
```

```python
data = df.copy()
cols = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Response']
df = df.drop(['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Response'],axis=1)
df
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
data_scaled = pd.DataFrame(scaler.transform(df),columns= df.columns)
data_scaled
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
pca.fit(data_scaled)
pca_df = pd.DataFrame(pca.transform(data_scaled), columns=(["col1","col2", "col3"]))
pca_df
```

# model building - clustering

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
elbow_g = KElbowVisualizer(KMeans(), k=10)
elbow_g.fit(pca_df)
elbow_g.show()
```

```python
inertias = []

for i in range(2,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(pca_df)
    inertias.append(kmeans.inertia_)

plt.plot(range(2,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

Taking 5 as the k value->>>>>

```python
x = pca_df["col1"]
y = pca_df["col2"]
z = pca_df["col3"]
kmeans = KMeans(n_clusters=5)
prediction = kmeans.fit_predict(pca_df)
pca_df["Clusters"] = prediction
df["Clusters"] = prediction
data["Clusters"] = prediction
fig = plt.figure(figsize=(12,10))
ax = plt.subplot(111, projection= "3d", label="bla")
ax.scatter(x,y,z, s=40, c=pca_df["Clusters"], marker="o", cmap="magma")
ax.set_title("The plot of the clusters")
plt.show()
```

So, we got 5 clusters

```python
plot = sns.countplot(x = df["Clusters"], palette = "Set3")
plot.set_title("Distribution Of The Clusters")
plt.show()
```

Above graph shows total count in each clusters

```python
scatter = sns.scatterplot(data = df, x = df["Spent"], y = df["Income"], hue = df["Clusters"], palette = "magma")
plt.title("Cluster Profile Based On Income And Spending")
plt.legend()
plt.show()
```

From above graph, we infer that , the customers belonging to the purple cluster has less income and spent less. while, the pink colored cluster has customers who has high income and spent more. yellow colored cluster has customers who has average income and and amount spent is on an average. the black cluster represents average income and spending is less to average. orange cluster represents average income and less spending.

```python
scatter = sns.scatterplot(data = df, x = df["total_days"], y = df["Spent"], hue = df["Clusters"], palette = "magma")
plt.title("Cluster Profile Based On total days And Spending")
plt.legend()
plt.show()
```
